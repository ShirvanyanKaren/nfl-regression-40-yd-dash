```python
blue, red = '#6CB4EE', '#DC143C'
years = [2018, 2019, 2020, 2021, 2022, 2023]
years_array = np.arange(1983, 2023 + 1)
nfl_data = nfl.import_seasonal_rosters(years)
years_array = years_array.tolist()
```

```python
wide_receivers = nfl_data[nfl_data['position']=='WR']
wide_receivers = wide_receivers[(wide_receivers['entry_year'] > 2000) & (wide_receivers['entry_year'] < 2018)] 
random_sample = wide_receivers.sample(n=300, axis=0)
random_sample
combine_results = nfl.import_combine_data(years_array, ['WR'])
combine_results_forty = combine_results
```

```python
players_forty = pd.merge(random_sample, combine_results_forty, on='player_name', how='left')
players_forty
```

```python
filtered_players = players_forty[players_forty['forty'].notna()]
filtered_players = filtered_players.sample(n=150, axis=0)
dict_list = filtered_players['player_name'].tolist()
dict_list
```

```python
filtered_players['player_name']
filtered_players
```

```python
player_stats = {}

for player in dict_list:
    player_name_with_dash = '-'.join(player.lower().split())
    url = f'https://www.nfl.com/players/{player_name_with_dash}/stats/'


    response = requests.get(url)

    if response.status_code == 200:
 
        soup = BeautifulSoup(response.content, 'html.parser')


        touchdowns_element = soup.find('th', {'aria-label': 'Stats.Receiving.Touchdowns'})
        receiving_yards_element = soup.find('th', {'aria-label': 'Stats.Receiving.Yards'})

        touchdowns = int(touchdowns_element.get_text(strip=True)) if touchdowns_element else None
        receiving_yards = float(receiving_yards_element.get_text(strip=True)) if receiving_yards_element else None

        player_stats[player] = {'touchdowns': touchdowns, 'receiving_yards': receiving_yards}
    else:
        print(f"Failed to retrieve data from {url}")


for player, stats in player_stats.items():
    print(f"{player}: {stats}")
```

```python
filtered_player = {}
for key in player_stats:
    if player_stats[key]['touchdowns'] != None:
        print(key)
        filtered_player[key] = {'touchdowns': player_stats[key]['touchdowns'],'receiving_yards': player_stats[key]['receiving_yards'] }
```

```python
filtered_players_df = pd.DataFrame.from_dict(filtered_player, orient='index')

filtered_players_df['player_name'] = filtered_players_df.index
```

```python
combined_df = pd.merge(filtered_players_df, filtered_players, on='player_name', how='left').set_index('player_name')
```

```python
combined_df.head(5)
```

```python
touchdown_ols = smf.ols(formula = 'touchdowns ~ forty', data = combined_df)
touchdown_regression = touchdown_ols.fit()
touchdown_regression.summary()

receiving_yards_ols = smf.ols(formula = 'receiving_yards ~ forty', data = combined_df)
receiving_yards_ols_reg = receiving_yards_ols.fit()
receiving_yards_ols_reg.summary()
```

```python
touchdown_regression.params.Intercept
touchdown_regression_rsq = "{:.4f}".format(touchdown_regression.rsquared)

touchdown_regression_formula = f'Y = {touchdown_regression.params["Intercept"]:.2f} + {touchdown_regression.params["forty"]:.2f}x'

print('Regression Formula:', touchdown_regression_formula)
print('R-Squared: ', touchdown_regression_rsq)


receiving_yards_ols_reg.params.Intercept
receiving_yards_ols_rsq = "{:.4f}".format(receiving_yards_ols_reg.rsquared)

receiving_yards_ols_reg_formula = f'Y = {receiving_yards_ols_reg.params["Intercept"]:.2f} + {receiving_yards_ols_reg.params["forty"]:.2f}x'

print('Regression Formula:', receiving_yards_ols_reg_formula)
print('R-Squared: ', receiving_yards_ols_rsq)
```

```python
plt.figure(figsize=(14, 10))
formula_box = dict(boxstyle='round', facecolor='white', edgecolor='black', alpha=0.7)


sns.regplot(x='forty', y='touchdowns', data=combined_df, color=blue, line_kws={'linestyle':'--', 'color': red}, scatter_kws={'s': 20})
sns.set_style('whitegrid')
plt.ylabel('Touch Downs', fontsize=14, labelpad=20)
plt.xlabel('Forty Time', fontsize=14, labelpad=20)
plt.title('Regression Plot: Touchdowns vs Forty Yard Dash Time', fontsize=16, pad=15)


annotation_text = f'R-Squared: {touchdown_regression_rsq}\nRegression Formula: {touchdown_regression_formula}'
plt.annotate(annotation_text, xy=(0.6, 0.85), xycoords='axes fraction', fontsize=14, color='dimgray', bbox=formula_box)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


plt.grid(linestyle='--', linewidth=0.5)


legend_unrate =['TDs', '40 Time']

plt.legend(labels=legend_unrate, fontsize='large')
plt.show()
```

```python
plt.figure(figsize=(14, 10))
formula_box = dict(boxstyle='round', facecolor='white', edgecolor='black', alpha=0.7)


sns.regplot(x='forty', y='receiving_yards', data=combined_df, color=blue, line_kws={'linestyle':'--', 'color': red}, scatter_kws={'s': 20})
sns.set_style('whitegrid')
plt.ylabel('Receiving Yards', fontsize=14, labelpad=20)
plt.xlabel('Forty Time', fontsize=14, labelpad=20)
plt.title('Regression Plot: Receiving Yards vs Forty Yard Dash Time', fontsize=16, pad=15)


annotation_text = f'R-Squared: {receiving_yards_ols_rsq}\nRegression Formula: {receiving_yards_ols_reg_formula}'
plt.annotate(annotation_text, xy=(0.6, 0.85), xycoords='axes fraction', fontsize=14, color='dimgray', bbox=formula_box)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


plt.grid(color='lightgray', linestyle='--', linewidth=0.5)

legend_unrate =['RYs', '40 Time']

plt.legend(labels=legend_unrate, fontsize='large')
plt.show()
```
